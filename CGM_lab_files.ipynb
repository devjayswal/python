```python
import matplotlib.pyplot as plt
import math

def rotate_line(x1, y1, x2, y2, angle_degrees):
    # Create a Matplotlib figure and axis
    fig, ax = plt.subplots()

    # Original line
    ax.plot([x1, x2], [y1, y2], color='blue', label='Original line')

    # Convert angle from degrees to radians
    angle_rad = math.radians(angle_degrees)

    # Calculate new coordinates after rotation
    x1_rotated = x1 * math.cos(angle_rad) - y1 * math.sin(angle_rad)
    y1_rotated = x1 * math.sin(angle_rad) + y1 * math.cos(angle_rad)
    x2_rotated = x2 * math.cos(angle_rad) - y2 * math.sin(angle_rad)
    y2_rotated = x2 * math.sin(angle_rad) + y2 * math.cos(angle_rad)

    # Rotated line
    ax.plot([x1_rotated, x2_rotated], [y1_rotated, y2_rotated], color='red', label='Rotated line')

    # Set axis limits
    ax.set_xlim(min(x1, x2, x1_rotated, x2_rotated) - 5, max(x1, x2, x1_rotated, x2_rotated) + 5)
    ax.set_ylim(min(y1, y2, y1_rotated, y2_rotated) - 5, max(y1, y2, y1_rotated, y2_rotated) + 5)

    # Add labels and legend
    ax.set_xlabel('X-Axis')
    ax.set_ylabel('Y-Axis')
    ax.set_title('Line Rotation\nDev Jayswal, Roll No: 0901AM221031')
    ax.legend()

    # Display the plot
    plt.grid()
    plt.show()

def main():
    x1, y1 = map(int, input("Enter the starting point of the line segment (x1 y1): ").split())
    x2, y2 = map(int, input("Enter the ending point of the line segment (x2 y2): ").split())
    angle_degrees = float(input("Enter the rotation angle in degrees: "))

    rotate_line(x1, y1, x2, y2, angle_degrees)

if __name__ == "__main__":
    main()
```

```python
import PIL
from PIL import Image, ImageDraw

import collections

def floodfill(x, y, color, image):
    source = image.getpixel((x, y))
    if source != color:
        pixels = collections.deque([(x, y)])
        while pixels:
            x, y = place = pixels.popleft()
            image.putpixel(place, color)
            for x_offset in -1, 1:
                x_offset += x
                for y_offset in -1, 1:
                    y_offset += y
                    new_place = x_offset, y_offset
                    if image.getpixel(new_place) == source:
                        pixels.append(new_place)

def main():
    # Create a new image object
    img = PIL.Image.new('RGB', (300, 300), (255, 255, 255))

    # Draw a triangle and fill it with red color
    triangle = [(50, 50), (150, 150), (50, 150)]
    PIL.ImageDraw.Draw(img).polygon(triangle, outline=(0, 0, 0), fill=(255, 0, 0))
    floodfill(75, 75, (255, 0, 0), img)

    # Draw a rectangle and fill it with green color
    rectangle = [(200, 50), (250, 50), (250, 100), (200, 100)]
    PIL.ImageDraw.Draw(img).polygon(rectangle, outline=(0, 0, 0), fill=(0, 255, 0))
    floodfill(225, 75, (0, 255, 0), img)

    # Draw a pentagon and fill it with blue color
    pentagon = [(50, 200), (100, 200), (125, 250), (75, 300), (25, 250)]
    PIL.ImageDraw.Draw(img).polygon(pentagon,outline=(0,0,0),fill=(0,0,255))
    floodfill(75 ,225 ,(0 ,0 ,255) ,img)

    # Save the image to disk
    img.save('polygons.png')
    img.show()

if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt

# Defining region codes
INSIDE = 0  # 0000
LEFT = 1    # 0001
RIGHT = 2   # 0010
BOTTOM = 4  # 0100
TOP = 8     # 1000

# Defining x_max, y_max and x_min, y_min for rectangle
x_max = 10.0
y_max = 8.0
x_min = 4.0
y_min = 4.0

# Function to compute region code for a point(x, y)
def compute_code(x, y):
    code = INSIDE
    if x < x_min:
        code |= LEFT
    elif x > x_max:
        code |= RIGHT
    if y < y_min:
        code |= BOTTOM
    elif y > y_max:
        code |= TOP
    return code

# Implementing Cohen-Sutherland algorithm
# Clipping a line from P1 = (x1, y1) to P2 = (x2, y2)
def cohen_sutherland_line_clipping(x1, y1, x2, y2):
    # Compute region codes for P1, P2
    code1 = compute_code(x1, y1)
    code2 = compute_code(x2, y2)
    accept = False

    while True:
        # If both endpoints lie within rectangle
        if code1 == 0 and code2 == 0:
            accept = True
            break
        # If both endpoints are outside rectangle
        elif (code1 & code2) != 0:
            break
        # Some segment of line lies within the rectangle
        else:
            # Line Needs clipping
            x = 1.0
            y = 1.0
            if code1 != 0:
                code_out = code1
            else:
                code_out = code2

            # Find intersection point
            if code_out & TOP:
                x = x1 + (x2 - x1) * (y_max - y1) / (y2 - y1)
                y = y_max
            elif code_out & BOTTOM:
                x = x1 + (x2 - x1) * (y_min - y1) / (y2 - y1)
                y = y_min
            elif code_out & RIGHT:
                y = y1 + (y2 - y1) * (x_max - x1) / (x2 - x1)
                x = x_max
            elif code_out & LEFT:
                y = y1 + (y2 - y1) * (x_min - x1) / (x2 - x1)
                x = x_min

            # Move the point outside the rectangle to intersection point
            if code_out == code1:
                x1, y1 = x, y
                code1 = compute_code(x1, y1)
            else:
                x2, y2 = x, y
                code2 = compute_code(x2, y2)

    if accept:
        return x1, y1, x2, y2
    else:
        return None

# Test the algorithm with line coordinates
x1, y1, x2, y2 = map(float, input("Enter coordinates (x1 y1 x2 y2): ").split())
clipped_line = cohen_sutherland_line_clipping(x1, y1, x2, y2)

# Plotting the line and rectangle
plt.figure()

# Before clipping
plt.subplot(1, 2, 1)
plt.plot([x1, x2], [y1, y2], color='blue', label='Original Line')
plt.plot([x_min, x_max, x_max, x_min, x_min], [y_min, y_min, y_max, y_max, y_min], color='red', label='Clipping Window')
plt.legend()
plt.title('Before Clipping')

# After clipping
plt.subplot(1, 2, 2)
if clipped_line:
    cx1, cy1, cx2, cy2 = clipped_line
    plt.plot([cx1, cx2], [cy1, cy2], color='green', label='Clipped Line')
plt.plot([x_min, x_max, x_max, x_min, x_min], [y_min, y_min, y_max, y_max, y_min], color='red', label='Clipping Window')
plt.legend()
plt.title('After Clipping')

plt.show()
print("Name :-Dev Jayswal \nEnroll no:- 0901AM221031")
```

```python
# Sutherland-Hodgman Polygon Clipping Algorithm
# Implemented by [Your Name]
# Enrollment Number: [Your Enrollment Number]

import matplotlib.pyplot as plt

# Define the clipping window coordinates
x_min, y_min = 50, 50
x_max, y_max = 150, 150

# Define the polygon vertices as a list of (x, y) tuples
polygon = [(75, 100), (100, 150), (125, 100), (100, 50)]

# Function to clip a polygon against a rectangular window
def sutherland_hodgman_clip(polygon, x_min, y_min, x_max, y_max):
    def inside(p, clip_edge):
        return (clip_edge[1][0] - clip_edge[0][0]) * (p[1] - clip_edge[0][1]) > (clip_edge[1][1] - clip_edge[0][1]) * (p[0] - clip_edge[0][0])

    def intersection(p1, p2, clip_edge):
        dx1, dy1 = p2[0] - p1[0], p2[1] - p1[1]
        dx2, dy2 = clip_edge[1][0] - clip_edge[0][0], clip_edge[1][1] - clip_edge[0][1]
        d = dx1 * dy2 - dx2 * dy1

        if d == 0:
            return None

        t = ((clip_edge[0][0] - p1[0]) * dy2 - (clip_edge[0][1] - p1[1]) * dx2) / d
        if 0 <= t <= 1:
            x = p1[0] + t * dx1
            y = p1[1] + t * dy1
            return (x, y)
        else:
            return None

    output = polygon

    for clip_edge in [((x_min, y_min), (x_max, y_min)), ((x_max, y_min), (x_max, y_max)), ((x_max, y_max), (x_min, y_max)), ((x_min, y_max), (x_min, y_min))]:
        input_polygon = output
        output = []

        prev = input_polygon[-1]

        for current in input_polygon:
            if inside(current, clip_edge):
                if not inside(prev, clip_edge):
                    intersection_point = intersection(prev, current, clip_edge)
                    if intersection_point:
                        output.append(intersection_point)
                output.append(current)
            elif inside(prev, clip_edge):
                intersection_point = intersection(prev, current, clip_edge)
                if intersection_point:
                    output.append(intersection_point)
            prev = current

    return output

# Clip the polygon
clipped_polygon = sutherland_hodgman_clip(polygon, x_min, y_min, x_max, y_max)

# Create subplots
plt.figure(figsize=(12, 5))

# Original Polygon (subplot 1)
plt.subplot(1, 2, 1)
x, y = zip(*polygon)
plt.plot(x, y, label='Original Polygon', color='blue')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Original Polygon')

# Clipped Polygon (subplot 2)
plt.subplot(1, 2, 2)
x, y = zip(*clipped_polygon)
plt.plot(x, y, label='Clipped Polygon', color='green')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Clipped Polygon')

plt.tight_layout()
plt.show()
```
